## Descripción y Justificación del método

Automated machine learning (AutoML) corresponde al proceso de aplicar machine lerning a problemas reales. Usualmente para elaborar una aplicación de machine learning, el desarrollador requiere conocer tenicas de pre procesamiento de datos, extracción de features y métodos de selección de features. Una vez habiendo realizado este proceso el desarrollor procede a realizar la elaboración del algoritmo realizando la selección de un método y la optimización de hiperparámetros, de forma que se obtenga la mayor precisión posible para el modelo. 

Una limitiación de este enfoque, es que para usuarios inexpertos muchos de estos pasos resultan significativamente complejos como para implementarlos. De modo que el AutoML es una propuesta alternativa. La idea sería automatizar el proceso de aplicar machine learning, generando soluciones más simples, generación fácil de soluciones y modelos que superen aquellos que fueron elaborados manualmente. 

De este modo con automated machine learning se obtienen modelos de alta precisión, con una reducción significativa de tiempo invertido en el desarrollo del modelo.


## Casos de uso

El autoML se vuelve una herramienta de gran utilidad para los siguientes casos:

- Implementar soluciones de machine learning sin implementar un conocimiento extensivo de programación
- Reducir costos de tiempo y recrusos
- Proveer soluciones de forma agil


Así mismo el autoML se puede implementar en las siguientes fases del proceso de machine learning

- Preparación automatizada de datos (tipos de datos, determinar si implementar regresión o clasificación)
- Automatización de extracción de features (manejo de datos faltantes y sesgos, selección y extracción de features) 
- Selección automatizada del pipeline (se considera limitaciones de tiempo y memoria)
- Análisis automatizado de los resultados obtenidos.

## Diferentes ejemplos (sabores) de AutoML

A continuación se exponen algunos ejemlpos de bibliotecas de autoML, se visualizaran parámetros como tiempo de duración y preicisión del método. Para todas estas se usó un set de datos para determinar si una persona posee o no una arritmia cardiaca.


Primeramente se hace un preprocesado de los datos, realizando una selección de las columnas que se usarán como hiperparámetros y cual es la salida que deseamos obtener. El conjuto de datos original calisificaba 16 tipos de arritmias, nuestro objetivo es simplificar el estudio un poco donde el valor de 1 en diagnosis representa una persona sana y un valor de 16 representa una persona con arritimia.

```python
import pandas as pd
data = pd.read_csv('arrythmia.csv')
data.head()
cols_to_use = ['q_wave','r_wave','s_wave',"R'_wave",'age','sex','height','weight','qrs_duration', 'p-r_interval','q-t_interval','t_interval','p_interval','qrs','heart_rate','diagnosis']
data_to_use = data[cols_to_use]
data_to_use.isnull().values.any()
data_to_use = data_to_use.dropna()
data_to_use.heart_rate = data_to_use.heart_rate.astype(int)
data_to_use.diagnosis = data_to_use.diagnosis.astype(int)
data_to_use.dtypes
data_to_use['diagnosis']=data_to_use['diagnosis'].replace({2:16,3:16,4:16,5:16,6:16,7: 16,8: 16,9: 16,10: 16,11: 16,12: 16,13: 16,14: 16,15: 16})
```

## Caso de referencia utilizando Bagging meta-estimator

Con el objetivo de utilizar un valor de referencia para comparar contra los métodos de autoML. Se procederá a correr el set de datos utilizando un método de ensamble. Este método utiliza otros classificadores como base en subconjuntos aleatorios del dataset original y luego genera un agreagado apartir de sus propias predicciones.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier  
from sklearn.metrics import accuracy_score

features = data_to_use[['q_wave','r_wave','s_wave',"R'_wave",'age','sex','height','weight','qrs_duration', 'p-r_interval','q-t_interval','t_interval','p_interval','qrs','heart_rate']]
labels = data_to_use['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state=1)
ExtraTreesClassifierModel = ExtraTreesClassifier(n_estimators=70, max_depth=None,min_samples_split=2, random_state=0)
ExtraTreesClassifierModel.fit(X_train, y_train)
bagging = BaggingClassifier(ExtraTreesClassifierModel, max_samples=0.9, max_features=0.9)
bagging.fit(X_train, y_train)
y_predict = bagging.predict(X_test)
accuracy_score(y_test, y_predict)
```

### Auto-sklearn

La primera biblioteca investigada es auto-sklearn, la cual es una biblioteca que utiliza la metodología Bayesiana para ajustar los hyperparámetros, los cuales van a alimentar los algoritmos tradicionales de machine learnining que están implementados dentro de scikit-learn.  

La instalación de la biblioteca es muy sencilla y solo requiere de la siguiente linea de código:

conda install gxx_linux-64 gcc_linux-64 swig
pip install auto-sklearn

```python
import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics

features = data_to_use[['q_wave','r_wave','s_wave',"R'_wave",'age','sex','height','weight','qrs_duration', 'p-r_interval','q-t_interval','t_interval','p_interval','qrs','heart_rate']]
labels = data_to_use['diagnosis']

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(features, labels, random_state=1)
automl = autosklearn.classification.AutoSklearnClassifier(n_jobs=8)
automl.fit(X_train, y_train)
y_hat = automl.predict(X_test, n_jobs=8)
print("Score of Accuracy", sklearn.metrics.accuracy_score(y_test, y_hat))
```

Una de las mayores venetjas observadas de la biblioteca fueron:

- uso de scikit-learn, la cual ha sido la biblioteca utilizada en el curso
- permite la implementación multicore, lo cual reduce los tiempos de ejecución
- significativa reducción del codigo implementado

Una desventaja notada con respecto a los resultados de la investigación 2 fue que a pesar de que se simplificó el set de datos, la precisión del modelo fue más baja. Una posible solución es utilizar los métodos de retrain para mejorar la precisión del modelo obtenido.

### Auto Keras

Auto-Keras es desarrollado por DATA Lab en la universidad de Texas A&M, esta biblioteca se le considera una alternativa a Google’s AutoML. Esto se debe a que está biblioteca también utiliza Neural Architecture Search (NAS), pero aplica “network morphism” (mantener la funcionalidad de la red mientras se modifica la arquitectura)juento con la optimización Bayesiana lo que de acuerdo a los desarrolladores permite una busqueda más eficiente de la red neuronal. 

La istalación de la biblioteca es muy sencilla solo requiere implementar las siguientes 2 lineas de código

conda install -c conda-forge keras tensorflow
pip install auto-keras


```python
import autokeras as ak 
import sklearn.model_selection


features = data_to_use[['q_wave','r_wave','s_wave',"R'_wave",'age','sex','height','weight','qrs_duration', 'p-r_interval','q-t_interval','t_interval','p_interval','qrs','heart_rate']]
labels = data_to_use['diagnosis']

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(features.to_numpy(),labels.to_numpy(),test_size=0.25,shuffle=True,random_state=1)
X_train = X_train.reshape(X_train.shape + (1,))
X_test = X_test.reshape(X_test.shape + (1,))
clf = ak.ImageClassifier(augment=True,resume=False,searcher_args={ 'trainer_args':{'max_iter_num':20}},verbose=True)
clf.fit(X_train, y_train, time_limit=(10*60*60))
clf.final_fit(X_train, y_train, X_test, y_test, retrain=False)
y=clf.evaluate(X_test, y_test)
print(y * 100)




```

# TPOT Light

**TPOT light** es un método de TPOT el cual restringe el uso computacional de los procesadores del sistema, además de los constructores de las características, los selectores, modelos y los parámetros que se van a utilizar para minimizar el error del módelo. Sólo se utilizan simples operadores que se ejecutan de manera sencilla a nivel de utilización computacional. 

Este método es muy putil para encontrar de manera rápida y simple un modelo que permita clasificar el problema.


```python
tpot_light = TPOTClassifier(generations=5, population_size=40, verbosity=2,config_dict='TPOT light')
```

```python
tpot_light.fit(X_train, y_train)
```

```python
print(tpot_light.score(X_test, y_test))
```

# TPOT MDR

**TPOT MDR** a diferencia del método anterior es un modelo multidmensional de reducción, es decir toma en cuenta factores en tercera dimensión para buscar los operadores para maximizar la predicción del modelo. Este modelo es utilizado especialmente en el campo de la genética para el estudio de asociación del genoma completo.

Es muy posible que este modelo se ejecute de manera mas lenta que el TPOT Light debido a que el consumo computacional es mayor, se recomienda utilizar procesadores de manera paralela para la ejecución de las instrucciones.

```python
tpot_mdr = TPOTClassifier(generations=5, population_size=40, verbosity=2,config_dict='TPOT MDR')
```

```python
tpot_mdr.fit(X_train, y_train)
```

```python
print(tpot_mdr.score(X_test, y_test))
```
